```python
# This function gives percentage of defaulter in total applicants for every unique label of the category.
def value_wise_per(df , col):
    df1 = pd.DataFrame(columns=['Value','Percentage of Defaulters'])
    for value in df[col].unique():
        count = df[(df[col] == value) & (df.TARGET == 1)].shape[0]
        total =  df[df[col] == value].shape[0]
        df1 = df1.append({"Value" : value  , 'Percentage of Defaulters' : (count * 100/total) }, ignore_index=True)
    return df1.sort_values('Percentage of Defaulters', ascending=False)
```

```python
# percentage of defaulter applicants who owns and does not own car(s)
value_wise_per(inp0 , 'FLAG_OWN_CAR')
```

```python
# percentage of defaulter applicants who have and does't have realty(s)
value_wise_per(inp0 , 'FLAG_OWN_REALTY')
```

Creating a function to return percentage of defaulter in total applicants having the same label of the category.

```python
# This function gives percentage of defaulter in total applicants having the same label of the category.
def value_wise_per(df , col):
    df1 = pd.DataFrame(columns=['Value','Percentage of Defaulters'])
    for value in df[col].unique():
        count = df[(df[col] == value) & (df.TARGET == 1)].shape[0]
        total =  df[df[col] == value].shape[0]
        df1 = df1.append({"Value" : value  , 'Percentage of Defaulters' : (count * 100/total) }, ignore_index=True)
    return df1.sort_values('Percentage of Defaulters', ascending=False)
```
